```python
df['DataValue']=pd.to_numeric(df['DataValue'],errors='correc')
df1=df[df['DataValue'].notnull()]
df2=df1[(df1['Topic']=='Oral Health')&(df1['DataValueType']=='Crude Prevalence')]
df_mod1=df2[['LocationDesc','QuestionID','DataValue','StratificationCategoryID1','StratificationID1']]
df_mod1=pd.get_dummies(df_mod1)
```

```python
x=df_mod1.drop('DataValue',1)
y=df_mod1['DataValue']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=101)
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
ypre=lr.predict(x_test)
ypre
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,ypre)
```
