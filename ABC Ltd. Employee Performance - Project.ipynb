# LabelEncoding to Convert String to Numerical Value

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df.Gender=le.fit_transform(df['Gender'])
df.EducationBackground=le.fit_transform(df['EducationBackground'])
df.MaritalStatus=le.fit_transform(df['MaritalStatus'])
df.EmpDepartment=le.fit_transform(df['EmpDepartment'])
df.EmpJobRole=le.fit_transform(df['EmpJobRole'])
df.BusinessTravelFrequency=le.fit_transform(df['BusinessTravelFrequency'])
df.Attrition=le.fit_transform(df['Attrition'])
df.EmpNumber=le.fit_transform(df['EmpNumber'])
df.OverTime=le.fit_transform(df['OverTime'])
df.Age_Group=le.fit_transform(df['Age_Group'])
```

```python
df1=df
```

# Defining X & Y

```python
x=df.drop('PerformanceRating',1)
y=df['PerformanceRating']
```

# Splitting Train & Test Data

```python
from sklearn.cross_validation import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=101)
```

# Fitting Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier(n_estimators=500)
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

# Accuracy

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
data=pd.DataFrame(rf.feature_importances_*100.0, columns=['Ginivalue Percentage'])
data.index=x.columns
```

```python
final=data.sort_values(['Ginivalue Percentage'],ascending=False)
final
```
